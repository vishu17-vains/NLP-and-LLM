```python
# lets start working with bigrams, trigrams and ngrams
from nltk.util import bigrams, trigrams, ngrams
```

```python
strings = "Artificial Intelligence refers to the intelligence of machines"
quotes_tokens = nltk.word_tokenize(strings)
quotes_tokens
```

```python
# checking length od tokenize words
len(quotes_tokens)
```

```python
# creating bigrams from tokenize text
quotes_bigram = list(nltk.bigrams(quotes_tokens))
quotes_bigram
```

```python
# creating trigrams from tokenize text
quotes_trigrams = list(nltk.trigrams(quotes_tokens))
quotes_trigrams
```

```python
# crteating n_grams
quotes_ngrams = list(nltk.ngrams(quotes_tokens, 4))
quotes_ngrams
```

```python
# creating ngrams of size 5 and 7
quotes_ngrams5 = list(nltk.ngrams(quotes_tokens,5))
quotes_ngrams5

```

```python
# ngram of size 7
quotes_ngrams7 = list(nltk.ngrams(quotes_tokens, 7))
quotes_ngrams7
```

APPLYING STEEMING USING DIFFERENT STEMERS

```python
# importing porter stemmer
from nltk.stem import PorterStemmer
```

```python
pst = PorterStemmer()
```

```python
# apply porterStemmer to different words
pst.stem("winning"), pst.stem("studies"), pst.stem("buying")
```

```python
# using lancaster stemmer
from nltk.stem import LancasterStemmer
lst = LancasterStemmer()
```

```python
# apply lancasterStemmer to different words
lst.stem("winning"), lst.stem("studies"), lst.stem("buying")
```

```python
# using snowballStemmer
from nltk.stem import SnowballStemmer
sbst = SnowballStemmer('english')
```

```python
# applying snowballStremer to different words
sbst.stem("winning"), sbst.stem("studies"), sbst.stem("buying")
```

```python
# lemmatization (word_root_form) using wordNetLammitizer
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
# lemmitize word
lemmatizer.lemmatize("cacti")
```

```python
# part of speech (POS) tagging
for token in AI_sent:
    print(nltk.pos_tag(word_tokenize(token)))
```

```python
# another sentence example for pos tyagging
sent2 = "my name is vishwajeet upadhyay"
sent2_tokens = word_tokenize(sent2)
sent2_tokens
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# List of words
words = ["Innovation", "Internet", "Software", "Hardware", "Networking", "Cybersecurity",
         "AI", "Machine Learning", "Cloud Computing", "Blockchain", "Data", "Digital",
         "Virtual Reality", "Augmented Reality", "IoT"]

# Combine words into a single string
text = ' '.join(words)

# Generate word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()

```
